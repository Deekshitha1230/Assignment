```python
Details = {'Name': ['John', 'Jane', 'Mike', 'Emily',None,'Ram','Jack'],
        'Salary': [25000, 30000, 28000, 32000,None,3000,2500],
        'exp(yrs)': [2,3,4,5,None,1,3]}
a=pd.DataFrame(Details)
print(a)
```

```python
print(a.isnull().sum())
```

```python
c=a['Salary'].mean()
c
```

```python
filledvalue=a['Salary'].fillna(c)
print(filledvalue)
```

```python
d=a['Salary'].mode()
d
```

```python
filledvalue=a['Salary'].fillna(d)
print(filledvalue)
```

```python
e=a['Salary'].median()
e
```

```python
filledvalue=a['Salary'].fillna(e)
print(filledvalue)
```

```python
filledvalue=a['Name'].fillna('Sita')
print(filledvalue)
```

```python
df=pd.read_csv('africa_recession.csv')
df
```

```python
q1=df['cda'].quantile(0.25)
q3=df['cda'].quantile(0.75)
iqr=q3-q1
iqr
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
print('lower=',lower_bound)
print('upper=',upper_bound)
```

```python
outliers=df[(df['cda']>lower_bound)|(df['cda']<upper_bound)]
print(outliers)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.boxplot(outliers['cda'])
plt.show()
```

```python
outliers=df[(df['cda']<lower_bound)|(df['cda']>upper_bound)]
print(outliers)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.boxplot(outliers['cda'])
plt.show()
```
